### Generando gráficos sobre el dataset de usuarios
Volvamos a incorporar el dataset de usuarios que vimos en la guía anterior 

```python
import pandas as pd
usuarios = pd.read_csv("https://raw.githubusercontent.com/informaticaort/intro-datascience/main/notebooks/python_guide/usuarios.csv")
```

En este caso vamos a graficar las alturas en función de la edad, pero usando Matplotlib, y con un gráfico de puntos.

```python
plt.plot(usuarios['age'], usuarios['height(cm)'], '.') #Grafica las alturas en función de las edades
plt.title('Altura por edades') #Título del gráfico
```

Así como este tendremos muchas posibilidades de generar gráficos. Es importante seguir los pasos ejemplo que se muestran en el sitio oficial. Nadie se sabe de memoria todas las instrucciones necesarias para generar por ejemplo, el siguiente gráfico:

```python
labels = 'A', 'B', 'C'
sizes = [15, 45, 40]
explode = (0.1, 0.1, 0.1) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

Lo importante no es conocer de memoria todas las líneas de código que realizan este gráfico de torta. Lo importante es entender lo que pasa para poder luego utilizarlo con mis propios datos. 

```python
porcentaje_mayores = len(usuarios[usuarios["age"] >= 18]) * 100 / len(usuarios)
porcentaje_menores = len(usuarios[usuarios["age"] < 18]) * 100 / len(usuarios)

labels = 'Mayores de Edad', 'Menores de Edad'
sizes = [porcentaje_mayores, porcentaje_menores]
explode = (0.1, 0.1) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

### Gráficos sobre el dataset de Spotify
Para los siguientes ejemplos vamos a incorporar otro archivo, en esta oportunidad será un historial de reproducción de Spotify de un usuario random. 

```python
spotify_historial = pd.read_csv("https://raw.githubusercontent.com/informaticaort/intro-datascience/main/notebooks/spotify_history/spotify_history.csv")
spotify_historial
```

Aprovechemos la oportunidad para aprender algo nuevo de Pandas. 
Con sort_values podemos ordenar los datos de un dataset. 

```python
mas_escuchadas = spotify_historial.copy()
mas_escuchadas.sort_values(by=['msPlayed'], inplace=True, ascending=False)
mas_escuchadas.head(20)
```

En este caso gracias al ordenamiento descendente que hicimos, podemos ver cuáles son las 20 canciones o archivos más reproducidos.

No está de más aclarar que no modificamos el Dataframe de historial de spotify original. Hicimos una copia del mismo con la función copy. 

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.bar(mas_escuchadas.head(20).trackName, mas_escuchadas.head(20).msPlayed, color='darkmagenta')
ax.set(title="Top 10 Canciones Escuchadas",xlabel="Artistas",ylabel="No. de milisegundos");
plt.xticks(rotation=75);
```

Quedó un poco raro el gráfico y es porque el dataset tiene el tiempo de reproducción en milisegundos, deberíamos modificar esto. No hay problema! Para esto vamos a utilizar dos funciones (que las vamos a crear nosotros con Python, para convertir el horario de milisegundos a horas)

```python
def obtener_minutos(milisegundos):
    return (milisegundos/60000)

mas_escuchadas["minPlayed"] = mas_escuchadas["msPlayed"].apply(obtener_minutos).round(3)
mas_escuchadas
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.bar(mas_escuchadas.head(20).trackName, mas_escuchadas.head(20).minPlayed, color='darkgreen')
ax.set(title="Top 10 Canciones Escuchadas",xlabel="Artistas",ylabel="No. de minutos");
plt.xticks(rotation=75);
```

## Wordcloud : Una forma amena de mostrar la información en una investigación

Una nube de palabras puede venir bien dentro de nuestra investigación, para mostrar los items más frecuentes dentro de un set de datos.

### Importación de librería

```python
from wordcloud import WordCloud 
```

### Generación de la nube de palabras
Primero debemos preparar los datos para la misma, en este caso vamos a ver cuántas canciones hay por artista en este historial de reproducción. La función **groupby** es esencial, dado que agrupando por el artista, cuenta la cantidad de registros en los cuales aparece cada uno. 

```python
artistas_favoritos = spotify_historial.groupby(["artistName"]).count()
artistas_favoritos.sort_values(by="msPlayed", ascending=False).head(100)
```

Luego procedemos a realizar la nube de palabras. Como siempre decimos, no es necesario conocer el 100% de esta instrucción. Sí, lo que tenemos que modificar (o queremos modificar) para personalizarla.

```python
fig, ax = plt.subplots(figsize=(15,10))
wordcloud = WordCloud(width=1000,height=600, max_words=50,relative_scaling=1,normalize_plurals=False, background_color="white").generate_from_frequencies(artistas_favoritos.trackName)
ax.imshow(wordcloud, interpolation='bilinear')
plt.axis(False);
```

Mirá qué bella manera de visualizar los artistas más escuchados por la persona que estamos analizando...
